```python
import requests
from bs4 import BeautifulSoup
import re
import json

def get_job_links(url, header_s):
    payload={
        'find_work_mode1': '2_1011_0001_0000',
        'find_work_mode2': '2_1011_0002_0000',
        's_find_work_mode1': '軟體／工程全部',
        's_find_work_mode2': 'MIS／網管全部',
        'order_by': 'm_date',
        'order_ascend': 'desc',
        'strrec': '20',
        'search_key_word': '',
        'search_type': 'job',
        'us_menu': '',
        'search_item': '1',
        'search_from': 'joblist',
        'job_show_type': 'L'
    }
    res = requests.post(url, headers=header_s, data=payload)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, "lxml")
    pg_selector = soup.select('#top_title > table.sift > tbody > tr > td[align]')[0].text
    total_pg_nums = int(re.findall('(\d{,3})頁', pg_selector)[0])
    print('total page number = %s'%total_pg_nums)
    
    links_list = []
    
    for pg_num in range(1, total_pg_nums+1):
        payload['strrec'] = str(pg_num*20)
        print('payload[strrec] = %s'%payload['strrec'])
        res = requests.post(url, headers=header_s, data=payload)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text, "lxml")
        select_list = soup.select('.box_detail_list ul li .jobname')
        print('this page has %s links'%str(len(select_list)))
        for num in range(len(select_list)):
            #print('[WAR] in element %s'%select_list[num].t)
            tmp = re.findall('^job_refer.+', select_list[num]['href'])
            if len(tmp)>0:
                tail = tmp[0]
                link = 'https://www.yes123.com.tw/admin/'+tail
                links_list.append(link)
                print('added link = %s'%link)
    print('total got links number = %s'%str(len(links_list)))
    return links_list
    
    
    
```

```python
def sort_list(input_list):
    res={}
    for word in input_list:
        if word in res:
            res[word] += 1
        else:
            res[word] = 1
    rank = [[word, res[word]]for word in res]
    result=sorted(rank, key=lambda x: x[1], reverse=True)
    return result
```
