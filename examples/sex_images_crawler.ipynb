```python
import requests
import json
from bs4 import BeautifulSoup

def get_images(url, filename):
    header_s = {
        'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
        'Accept-Encoding':'gzip, deflate',
        'Accept-Language':'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6',
        'Cache-Control':'no-cache',
        'Connection':'keep-alive',
        'Host':'img1.mm131.me',
        'Pragma':'no-cache',
        'Referer':'http://img1.mm131.me/pic/3817/1.jpg',
        'Upgrade-Insecure-Requests':'1',
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Safari/537.36'
    }
    img_data = requests.get(url, headers=header_s).content
    with open(filename, 'wb') as handler:
        handler.write(img_data)
        
```

```python
def get_imgurl(url):
    try:
        res = requests.get(url, headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Safari/537.36'})
        soup = BeautifulSoup(res.text, "lxml")
        target = soup.select_one(".content-pic img")['src']
    except:
        target = ""
    return target
```
